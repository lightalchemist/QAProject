## Sort Dan Roth's dataset to match order of Jacana's dataset

---

```python
dr_train_datapath = "../data/QuestionType/train-less-than-40.output"
dr_test_datapath = "../data/QuestionType/test-less-than-40.output"

jacana_train_datapath = "../data/answerSelectionExperiments/data/train-less-than-40.xml"
jacana_test_datapath = "../data/answerSelectionExperiments/data/test-less-than-40.xml"
```

Associate each Dan Roth's "hashed" sentence with its original sentence

```python
def get_dr_data(filepath):
    data = {}
    with open(filepath) as infile:
        for line in infile:
            qn_type, sentence = line.strip().split(' ', 1)
            key = sentence.strip().replace(' ', '').lower()  # Remove whitespaces
            value = line
            data[key] = value
            
    return data
```

```python
# Get a question and its candidate answers
def get_QA_group(infile):
    question = []
    answers = []
    line = infile.readline().strip()
    if line == "":
        return None
    
    if not line.startswith("<QApairs"):
        raise Exception("Invalid data format: {}<-----".format(line))
    
    while not line.strip().startswith("</QApairs"):
        line = infile.readline().replace('\t', ' ')
        if line.strip().lower().startswith("<question"):
            line = infile.readline().replace('\t', ' ')
            question.append(line.strip())
        elif line.strip().lower().startswith("<positive"):
            line = infile.readline().replace('\t', ' ')
            answers.append(("positive", line.strip()))
        elif line.strip().lower().startswith("<negative"):
            line = infile.readline().replace('\t', ' ')
            answers.append(("negative", line.strip()))
    
    return {"question": question, "answers": answers}                
```

```python
def sort_dr_question_type_file(jacana_input_file, danroth_data_input_file, output_file):    
    dr_data = get_dr_data(danroth_data_input_file)    
    with open(jacana_input_file) as infile, \
             open(output_file, 'w') as outfile:

            while infile:
                group = get_QA_group(infile)

                # Check for EOF
                if group is None:
                    break

                # Extract question vector
                question = group["question"][0]
                key = question.strip().replace(' ', '').lower()

                assert(key in dr_data)
                dr_sentence = dr_data[key]

                print(dr_sentence.strip())
                outfile.write(dr_sentence.strip() + '\n')
```

```python
sort_dr_question_type_file(jacana_train_datapath, dr_train_datapath, "sorted-dr-train-less-than-40.txt")
sort_dr_question_type_file(jacana_test_datapath, dr_test_datapath, "sorted-dr-test-less-than-40.txt")
```
